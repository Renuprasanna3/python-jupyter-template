### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', s=100)  # s controls point size

# Add title and labels
plt.title('Scatter Plot with Annotations', fontsize=14, pad=20)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Annotate each point with its (x, y) value
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', 
                 (xi, yi), 
                 textcoords="offset points",
                 xytext=(0, 10),  # Position text above the point
                 ha='center',
                 fontsize=10)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
