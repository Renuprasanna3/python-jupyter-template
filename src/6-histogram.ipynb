### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.python
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create histogram with customizations
plt.figure(figsize=(8, 6))
plt.hist(data, 
         bins=5,  # Number of bins matching our data range
         color='#4CAF50',  # Green color
         edgecolor='black',  # Black edges for bars
         alpha=0.8)  # Slight transparency

# Add title and labels
plt.title('Frequency Distribution of Values', fontsize=14, pad=20)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Customize x-axis ticks to show all integer values
plt.xticks(range(1, 6))

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
