### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Custom colors
explode = (0.1, 0, 0, 0)  # Explode the Python slice

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors,
        autopct='%1.1f%%',  # Show percentages with 1 decimal place
        shadow=True, 
        startangle=140)

# Add title
plt.title('Programming Language Popularity', fontsize=14, pad=20)

# Equal aspect ratio ensures pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.tight_layout()
plt.show()
```
